```python
df = pd.read_csv("dataset_performance.csv")
print(df)
```

```python
df=df.drop(columns=['netflix_hours'])
print(df)
```

```python
df.info()
```

```python
updated_age = df[df['age']>=18]
print(updated_age)
```

```python
sorted_age = df.sort_values(by='age')
print(sorted_age)
```

```python
grouped_df=df.groupby('age')['study_hours_per_day'].mean()
print(grouped_df)
```

```python
df.rename(columns={'major':'major_subject'}, inplace=True)
print(df)
```

```python
df['exam_result']=df['exam_score'].apply(lambda x:'PASS' if x>=45 else'FAIL')
print(df)
```

```python
def categorize_age(Years):
    if Years<18:
        return 'child'
    else:
        return 'Adult'

df['Age_category'] = df['age'].apply(categorize_age)
print(df)
```

```python
A = df["major_subject"]
B = df["study_hours_per_day"]
plt.plot(A,B)
plt.title("Line Graph")
plt.xlabel("Gender")
plt.ylabel("study_hours_per_day")
plt.show()
```

```python
avg_scores = df.groupby('major_subject')['exam_score'].mean().sort_values()
plt.figure(figsize=(8, 5))
plt.plot(avg_scores.index, avg_scores.values, color='green')
plt.title('Average Exam Score by Major')
plt.xlabel('Major Subject')
plt.ylabel('Mean Exam Score')
plt.show()
```

```python
avg_scores = df.groupby('major_subject')['exam_score'].mean().sort_values()
plt.figure(figsize=(12, 6))
plt.bar(avg_scores.index, avg_scores.values, color='skyblue', edgecolor='yellow')
plt.title('Average Exam Score by Major (Bar)')
plt.xlabel('Major Subject')
plt.ylabel('Mean Exam Score')
plt.show()
```

```python
avg_scores = df.groupby('major_subject')['exam_score'].mean().sort_values()
plt.figure(figsize=(8, 5))
plt.barh(avg_scores.index, avg_scores.values, color='lightgreen', edgecolor='darkgreen')
plt.title('Average Exam Score by Major (Horizontal)')
plt.xlabel('Mean Exam Score')
plt.ylabel('Major Subject')
plt.show()
```

```python
avg_scores = df.groupby('major_subject')['exam_score'].mean().sort_values()
plt.figure(figsize=(8, 5))
plt.scatter(df['study_hours_per_day'], df['exam_score'], color='purple')
plt.title('Exam Score vs. Study Hours')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
```

```python
avg_scores = df.groupby('major_subject')['exam_score'].mean().sort_values()
plt.figure(figsize=(8, 5))
plt.hist(df['exam_score'], bins=20, color='orange', edgecolor='green')
plt.title('Distribution of Exam Scores')
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```

```python
major_counts = df['major_subject'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(major_counts.values, labels = major_counts.index,autopct='%1.1f%%')
plt.title('Major Subject Distribution')
plt.show()
```
